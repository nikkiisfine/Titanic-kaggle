```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv('titanic/train.csv')
test=pd.read_csv('titanic/test.csv')
print('训练数据集',train.shape,'测试训练集',test.shape)
```

```python
full=train.append(test,ignore_index=True)
print('合并后的数据集:',full.shape)
```

```python
full.head(5)
```

```python
full.describe()
```

```python
full.info()
```

```python
#处理float数据缺失   ???有待深一步学习
print('处理前:')
full.info()
#年龄age
ageSer=full.groupby('Sex').Age.mean()
full.set_index('Sex',inplace=True)
full['Age'].fillna(ageSer,inplace=True)
full.reset_index(inplace=True)
#船票价格
full['Fare']=full['Fare'].fillna(full['Fare'].mean())
print('处理后:')
full.info()
```

```python
# 处理字符串类型数据缺失
full['Embarked'].value_counts()
```

```python
full['Embarked']=full['Embarked'].fillna('S')
```

```python
full['Cabin']=full['Cabin'].fillna('U')
```

```python
full.info()
```

```python
#特征工程
```

```python
#年龄类别
#少年：年龄<=15
#青年：15-30
#中年：30-50
#老年：58-80
ageDf=pd.DataFrame()
bins=[0,15,30,50,80]
groupName=['Youth','YouthAdult','MiddleAged','Senior']
cat=pd.cut(full['Age'],bins,labels=groupName)
cat.value_counts()
```

```python
ageDf=pd.get_dummies(cat)
ageDf.head()
```

```python
sex_dict={'male':1,
         'female':0}
#map是指对Series里的每个数据都应用自定义的函数计算
full['Sex']=full['Sex'].map(sex_dict)
full['Sex'].head()
```

```python
#存放提取后的特征
EmbarkedDf=pd.DataFrame()
#使用get_dummies进行one-hot编码，产生虚拟变量（dummy variables），列名前缀是Embarked
EmbarkedDf=pd.get_dummies(full['Embarked'],prefix='Embarked')
EmbarkedDf.head()
```

```python
full=pd.concat([full,EmbarkedDf],axis=1)
```

```python
#客舱等级
PclassDf=pd.DataFrame()
PclassDf=pd.get_dummies(full['Pclass'],prefix='Pclass')
PclassDf.head()
```

```python
full['Name'].head()
```

```python
#定义提取名称头衔的函数
def getTitle(Name):
    str1=Name.split(',')[1]
    str2=str1.split('.')[0]
    str3=str2.strip()
    return str3
```

```python
titleDf=pd.DataFrame(dtype=int)
titleDf['Title']=full['Name'].map(getTitle)
titleDf.head()
```

```python
titleDf['Title'].value_counts()
```

```python
title_mapDict={
    'Capt':'Officer',
    'Col':'Officer',
    'Major':'Officer',
    'Jonkheer':'Royalty',
    'Don':'Royalty',
    'Sir':'Royalty',
    'Dr':'Officer',
    'Rev':'Officer',
    'the Countess':'Royalty',
    'Dona':'Royalty',
    'Mme':'Mrs',
    'Mlle':'Miss',
    'Ms':'Mrs',
    'Mr':'Mr',
    'Mrs':'Mrs',
    'Miss':'Miss',
    'Master':'Master',
    'Lady':'Royalty'
    
}
titleDf['Title']=titleDf['Title'].map(title_mapDict)
titleDf=pd.get_dummies(titleDf['Title'])
titleDf.head()
```

```python
#客舱号
full['Cabin'].head()
```

```python
cabinDf=pd.DataFrame(dtype=int)
full['Cabin']=full['Cabin'].map(lambda c:c[0])
cabinDf=pd.get_dummies(full['Cabin'],prefix='Cabin')
cabinDf.head()
```

```python
#船票编码
```

```python
#家庭
#小家庭Family_Single：家庭人数=1
#中等家庭Family_Small: 2<=家庭人数<=4
#大家庭Family_Large: 家庭人数>=5
#家庭人数（Family）=同代直系亲属数（Parch）+不同带直系亲属数（SibSp）+乘客自己
```

```python
familyDf=pd.DataFrame()
familyDf['FamilySize']=full['Parch']+full['SibSp']+1
familyDf['Family_Single']=familyDf['FamilySize'].map(lambda s:1 if s==1 else 0)
familyDf['Family_Small']=familyDf['FamilySize'].map(lambda s:1 if 2<=s<=4 else 0)
familyDf['Family_Large']=familyDf['FamilySize'].map(lambda s:1 if s>=5 else 0)
familyDf.head()
```

```python
full=pd.concat([full,PclassDf],axis=1)
full=pd.concat([full,familyDf],axis=1)
full=pd.concat([full,cabinDf],axis=1)
full=pd.concat([full,titleDf],axis=1)
full=pd.concat([full,ageDf],axis=1)
```

```python
#相关性矩阵
corrDf=full.corr()
corrDf
```

```python
corrDf['Survived'].sort_values(ascending=False) #如何快速选择绝对值较大的特征
```

```python
full_X=pd.concat([titleDf,#头衔
                 PclassDf,#客舱等级
                 familyDf,#家庭大小
                 full['Fare'],#船票价格
                 #cabinDf,#船舱号
                 EmbarkedDf,#登船港口
                 #full['Sex'], #性别 
                ageDf], axis=1)
full_X.head()                 
```

```python
#开始建模咯
```

```python
sourceRow=891
source_X=full_X.loc[0:sourceRow-1,:]
source_Y=full.loc[0:sourceRow-1,'Survived']
pred_X=full_X.loc[sourceRow:,:]
print('原始数据集有多少行', source_X.shape[0])
print('预测原始数据集有多少行', pred_X.shape[0])
```

```python
from sklearn.cross_validation import train_test_split

#建立模型用的训练数据集和测试数据集
#等号前面的顺序很重要，train_X&test_X train_test_split source_X
train_X,test_X,train_Y,test_Y=train_test_split(source_X,source_Y,train_size=0.8) 
#输出数据集大小
print('原始数据集特征',source_X.shape,
      '训练数据集特征',train_X.shape,
      '测试数据集特征',test_X.shape)
print('原始数据集标签',source_Y.shape,
      '训练数据集标签',train_Y.shape,
      '测试数据集标签',test_Y.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(train_X,train_Y)
```

```python
model.score(test_X,test_Y)
```

```python
pred_Y=model.predict(pred_X)
pred_Y=pred_Y.astype(int)
passenger_id=full.loc[sourceRow:,'PassengerId']
predDf=pd.DataFrame({
    'PassengerId':passenger_id,
    'Survived':pred_Y
})
predDf.shape
predDf.head()
```

```python
predDf.to_csv('titanic/pred.csv',index=False)
```
